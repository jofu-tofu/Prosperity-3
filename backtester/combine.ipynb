```python
import json
from datamodel import Trade
import util
import pandas as pd

sblogs, bk, mkt1 = util._parse_data('vwap_d1.log')
_, _, mkt2 = util._parse_data('vwap_d2.log')
_, _, mkt3 = util._parse_data('vwap_d3.log')

concat = pd.concat([mkt1, mkt2, mkt3], axis=0)

output = ""
output += "Sandbox logs:\n"
sblogs = [{
                    "sandboxLog": row.sandboxLog,
                    "lambdaLog": row.lambdaLog,
                    "timestamp": row.Index,
                } for row in sblogs.itertuples()]
for i in sblogs:
    output += json.dumps(i, indent=2) + "\n"

output += "\n\n\n\nActivities log:\n"
market_data_csv = bk.to_csv(sep=";")
market_data_csv = market_data_csv.replace("\r\n", "\n")
output += market_data_csv

output += "\n\n\n\nTrade History:\n"
trades = [{
            "timestamp": trade.Index,
            "buyer": trade.buyer,
            "seller": trade.seller,
            "symbol": trade.symbol,
            "currency": "SEASHELLS",
            "price": trade.price,
            "quantity": trade.quantity,
        } for trade in concat.itertuples()]
output += json.dumps(trades, indent=2)

output

with open("data/combinedvwap.log", "w") as file:
    file.write(output)
```
